```python
def get_detection_image(input_name):
  fa = face_alignment.FaceAlignment(face_alignment.LandmarksType._2D, device='cpu', face_detector='sfd')
  input = io.imread(input_name)
  det = fa.get_landmarks(input)
  x=[]
  y=[]
  det=det[0]
  for detection in det:
    #print(detection)
    x.append(detection[0])
    y.append(detection[1])
  plt.scatter(x,y)
  plt.show()

get_detection_image('소녀시대 태연_img_116.jpg')
get_detection_image('/content/보아1.jpg')



# plt.imshow(input)
# x=[]
# y=[]
# det=det[0]
# for detection in det:
#   print(detection)
#   x.append(detection[0])
#   y.append(detection[1])
# plt.scatter(x,y)
```
